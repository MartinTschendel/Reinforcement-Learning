```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

<img src="UCB_Algorithm_Slide.png" width=600 align="left">

```python
#N (as argumnet of the function, is the number of people (rounds)
def get_best_ad(N):

    #number of ads
    d = 10
    #creation of list that contains selected ads
    ads_selected = []

    #number of times the ad i was selected up to round n
    #list with 10 zeros
    numbers_of_selections = [0] * d
    #sum of rewards of the ad i up to round n
    sums_of_rewards = [0] * d
    total_reward = 0

    import math


    # n stands for the round
    for n in range(0, N):
        ad = 0
        max_upper_bound = 0
        #second for loop goes through the 10 different ads in one round
        for i in range(0, d):
            #check if ad(i) was already selected
            if (numbers_of_selections[i] > 0):
                average_reward = sums_of_rewards[i] / numbers_of_selections[i]
                #if n = 0, log(n) would be -infinitive, so we start with n + 1 
                delta_i = math.sqrt(3/2 * math.log(n + 1) / numbers_of_selections[i])
                #upper bound (UCB) of confidence interval
                upper_bound = average_reward + delta_i
            #select the ad i that has the maximum UCB
            #the else statement deals with the ad that has not been selected
            else:
                #give this ad a very high UCB
                upper_bound = 1e400
            #update the max_upper_bound
            if upper_bound > max_upper_bound:
                max_upper_bound = upper_bound
                #select the ad with th highest UCB
                ad = i
            #updating variables
            ads_selected.append(ad)
            numbers_of_selections[ad] += 1
            #n refers to the round (customer), and ad refers to the specific adverstisment
            #if customer clicks on ad, 1 will be added
            reward = dataset.values[n, ad]
            sums_of_rewards[ad] += reward
            total_reward += reward
    return ads_selected
#the function above returns the variable ads_selected, what is needed to
#plot the histogram below
```

```python

plt.hist(get_best_ad(10000))
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
